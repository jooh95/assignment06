```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define variables for noisy, clean data.

```python
num     = 201
std     = 20
a       = 2 
b       = 10
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
```

# Concate one vector and X vector.
  \begin{equation}X=\begin{bmatrix}
         1, X_{1} \\
         1, X_{2} \\
         \vdots\\
    \vdots\\
         1, X_{n} \\
        \end{bmatrix}
          \end{equation}

```python
ones = np.ones(201)

n_x = np.array([ones, x]).T
```

# Calculate the left pseudo inverse.
\begin{equation}
I = (A^{T}A)^{-1}A^{T}
\end{equation}

```python
s_inverse = np.dot(np.linalg.inv(np.dot(n_x.T, n_x)), n_x.T)
```

# Define noisy and clean data.

```python
y1      = a * x + nn * std + b

y2      = a * x + b
```

# Calculate the line approximated through least square. 

```python
p = np.dot(s_inverse, y1)

y3 = p[0] * x + p[1]
```

# Visulaize the result graphs.

```python
plt.title('Noisy Data')
plt.plot(x, y1, 'r.')
plt.show()

plt.title('Clean Data')
plt.plot(x, y2, 'g.')
plt.show()

plt.title('Least Square approximated Line')
plt.plot(x, y3, 'b--')
plt.show()
```
